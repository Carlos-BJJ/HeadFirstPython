```python
#Importei os módulos necessarios
import os
import swimclub

#Listei os arquivos do diretorio swimdata/ que é a constante FOLDER
swim_files = os.listdir(swimclub.FOLDER)

#Removo o arquivo indesejado da lista
swim_files.remove('.DS_Store')

#Verifico se agora está realmente somente os 60 arquivos dos nadadores
print(len(swim_files))
```

```python
#Crio uma lista vazia chamada "nadadores"
swimmers = []

#Crio um loop for para processar os arquivos da lista e criar uma lista nova só com os nomes dos nadadores
for s in swim_files:
   #O zero acessa o primeiro indice da tupla (nome, categoria, distancia, estilo, tempos, media_formatada, convercoes)
   swimmers.append(swimclub.read_swim_data(s)[0])

print(swimmers)
```

```python
print(len(swimmers))
```

```python
#Utilizo a BIF list para transformar a tupla em lista, e set é para remover os nomes duplicados
names = list(set(swimmers))
```

```python
print(sorted(names))
```

```python
print(len(list(set(swimmers))))
```

```python
swimmers = []

#Uso o loop para remover os itens duplicados da tupla e coloca-los na lista de nadadores
for file in swim_files:
    name = swimclub.read_swim_data(file)[0]
    if name not in swimmers:
        swimmers.append(name)
print(swimmers)
```

```python
print(sorted(swimmers))
```
