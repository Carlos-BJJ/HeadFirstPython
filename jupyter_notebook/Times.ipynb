```python
FN = "Darius-13-100m-Fly.txt"
```

```python
nome, categoria, distancia, estilo = FN.removesuffix(".txt").split("-")
```

```python
FOLDER = "swimdata/"
```

```python
with open(FOLDER + FN) as arquivo:
    linhas = arquivo.readlines()
    tempos = linhas[0].strip().split(",")
```

```python
convercoes = []
for t in tempos:
    minutos, resto = t.split(":")
    segundos, centesimos = resto.split(".")
    int_conversao = (int(minutos) * 60 * 100) + (int(segundos) * 100) + (int(centesimos))
    convercoes.append(int_conversao)
```

```python
import statistics

media_convercoes = statistics.mean(convercoes)
min_seg, cent = str(round(media_convercoes / 100, 2)).split(".")
min = int(min_seg) // 60
seg = int(min_seg) - min * 60
media_convercoes = str(min) + ":" + str(seg) + "." + cent
```

```python
print("Nome:",nome)
print("Categoria: Sub -",categoria)
print("Distância:",distancia)
print("Estilo de Nado:",estilo)
print("\nA lista de tempos do", nome, "é:\n" ,tempos)
print("\nO tempo médio do", nome, "é:\n" ,media_convercoes)
```
