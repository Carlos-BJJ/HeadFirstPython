```python

#Importo o módulo que vai ser usado
import swimclub

#Identifico qual nome do arquivo a função vai ser buscado no diretorio
fn = "Darius-13-100m-Fly.txt"

#Descompacto as variaveis que vou utilizar e uso o double dunder com "*" para atribuir o restante das variaveis
(swimmer, age, distance, stroke, *__) = swimclub.read_swim_data(fn)

#Crio uma variavel que será o titulo do código em HTML (#Usando um método menos eficaz para transformar variaveis em strings)
title = swimmer + " (Under " + age + ") " + distance + " " + stroke
```

```python
title
```

```python
#Testo um método menos eficaz para transformar variaveis em strings
html = " <!DOCTYPE <html><html><head><title> " + title
html = html + " </title></head><body><h3> " + title + " </h3>"
```

```python
html
```

```python
#Agora uso f'string, uma maneira mais eficiente de declarar variaveis em uma string
title = f"{swimmer} (Under {age}) {distance} {stroke}"
```

```python
title
```

```python
#Agora uso f'string novamente, mas já com o código completo
html = f"""<!DOCTYPE html>
<html>
    <head>
        <title>
            {title}
        </title>
    </head>
    <body>
        <h3>{title}</h3>"""
```

```python
#Imprimo a variavel html
print(html)
```
