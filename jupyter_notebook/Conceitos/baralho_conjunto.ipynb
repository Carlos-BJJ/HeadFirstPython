```python
baralho = set()
```

```python
type(baralho)
```

```python
len(baralho)
```

```python
naipes = ["Clubs", "Spades", "Hearts", "Diamonds"]
faces = ["Jack", "Queen", "King", "Ace"]
numbered = [2, 3, 4, 5, 6, 7, 8, 9, 10]
```

```python
for naipe in naipes:
    for card in faces + numbered:
        baralho.add((card, "of", naipe))
```

```python
len(baralho)
```

```python
import random

carta = random.choice(list(baralho))
print(carta)
```

```python
baralho.remove(carta)
len(baralho)
```

```python
def draw():
    carta = random.choice(list(baralho))
    baralho.remove(carta)
    return carta
```

```python
print(draw())
```

```python
len(baralho)
```

```python
if carta in baralho:
    print("A carta está no baralho!!!")
else:
    print("A carta não está no baralho!!!")
```
