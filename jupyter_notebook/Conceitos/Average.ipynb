```python
converts = []
for time in times:
    minutos, resto = time.split(":")
    segundos, centesimos = resto.split(".")
    int_conversao = (int(minutos) * 60 * 100) + (int(segundos) * 100) + (int(centesimos))
    converts.append(int_conversao)
```

```python
soma = 0
for t in converts:
    soma = soma + t
    media = soma / len(converts)
print(media)

#Cálculo de média sem uso da Biblioteca statistics e com a função mean
```

```python
import statistics
statistics.mean(converts)

#Cálculo de média com uso da Biblioteca statistics e com a função mean
```

```python
average = statistics.mean(converts)

round(average / 100, 2)

#Converti os centesimos em segundos arredondando a parte decimal para duas casas com a BIF round
```
