```python
import random

naipes = ["Clubs", "Spades", "Hearts", "Diamonds"]
faces = ["Jack", "Queen", "King", "Ace"]
numbered = [2, 3, 4, 5, 6, 7, 8, 9, 10]
```

```python
def draw():
    the_naipe = random.choice(naipes)
    the_card = random.choice(faces + numbered)
    return the_card, "of", the_naipe
```

```python
for _ in range(5):
    print(draw())
```
