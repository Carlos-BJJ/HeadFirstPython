```python
#Importo o módulo que vai ser usado
import swimclub
import hfpy_utils

#Identifico qual nome do arquivo a função vai ser buscado no diretorio
fn = "Darius-13-100m-Fly.txt"

#Descompacto as variaveis que vou utilizar e uso o double dunder com "*" para atribuir o restante das variaveis
(swimmer, age, distance, stroke, *__) = swimclub.read_swim_data(fn)

#Uso f'string, uma maneira mais eficiente de declarar variaveis em uma string
title = f"{swimmer} (Under {age}) {distance} {stroke}"

#Agora uso f'string novamente, mas já com o código completo
html = f"""<!DOCTYPE html>
<html>
    <head>
        <title>
            {title}
        </title>
    </head>
    <body>
        <h3>{title}</h3>"""
```

```python
#Descompacto as variaveis que vou utilizar e uso o double dunder com "*" para atribuir o restante das variaveis
*__, times, average, converts = swimclub.read_swim_data(fn)

#Crio uma variavel para guardar o maior valor da lista de converções
from_max = max(converts)

#Crio uma variavel vazia que irá ser substituida pelo código em HTML
svgs = ""

#Loop para utilizar a função do módulo "hfpy_utils" para a converção dos valores para o intervalo desejado (0 a 400)
for n,t in enumerate(times):
    bar_windth = hfpy_utils.convert2range(converts[n], 0, from_max, 0, 400)
    svgs = svgs + f"""
                    <svg height="30" width="400">
                        <rect height="30" width="{bar_windth}" style="fill:rgb(0,0,255);" />
                    </svg> {t} <br />
                  """   
```

```python
#Crio uma variavel para escrever o rodapé do código em HTML
footer = f"""
        <p>Average time: {average}</p>
    </body>
</html>
"""
```

```python
#Junto as 3 partes do código
page = html + svgs + footer
```

```python
#Imprimo pra ver se está tudo correto
print(page)
```
